```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,8))
x=np.array([20,25,35,55,65,75])
y=np.array([16,36,48,62,68,78])
z=np.array([8.8,19.7,18.6,19.2,20.3,15.4])
colors = ['green','lightgreen','brown','dodgerblue','yellow','red']

plt.xlabel("age of rider",size=20)
plt.text(20,85,"size of bubble=no of divers in that age group in (millions)",size=15)

plt.ylabel("death expectancy",size=20)
plt.gca().set_facecolor('lightyellow')
plt.scatter(x, y, s=z*300,c=colors,alpha=0.5)
plt.legend()
plt.xticks(np.arange(10,100,10))
plt.yticks(np.arange(20,100,10))
plt.title('death expectency of riders at speed above 60kmph according to age',size=20)
plt.show()
```
